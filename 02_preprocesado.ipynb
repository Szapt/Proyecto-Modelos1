### Cleaning 'E_VALORMATRICULAUNIVERSIDAD' column

Ahora convertiremos las descripciones textuales de los rangos monetarios en representaciones numéricas comparables. Para esto, tomaremos el promedio de los extremos del intervalo. Luego, debido a que los valores provienen de categorías, llenaremos los nulos con la moda.

Encoding de variables categóricas

```python
z_copy['E_VALORMATRICULAUNIVERSIDAD'].unique()
```

```python
z_copy['E_VALORMATRICULAUNIVERSIDAD'].head(15)
```

```python
import re
import numpy as np
import pandas as pd
import unicodedata

def convertir_rango_a_numero(valor):
    if pd.isna(valor):
        return np.nan

    texto = str(valor).lower()
    texto = ''.join(c for c in unicodedata.normalize('NFD', texto) if unicodedata.category(c) != 'Mn')

    # Caso 'No pagó matrícula'
    if "no pago" in texto:
        return 0.0

    # Buscar todas las apariciones de números con su contexto
    patrones = re.findall(r'(\d+\.?\d*)\s*(millon(?:es)?|mil)?', texto)
    numeros = []

    for num, unidad in patrones:
        n = float(num)
        if "millon" in unidad:
            n *= 1_000_000
        elif "mil" in unidad:
            n *= 1_000
        numeros.append(n)

    # Caso 'Entre X y menos de Y'
    if len(numeros) == 2:
        return np.mean(numeros)


    # Caso 'Menos de X'
    if "menos de" in texto and len(numeros) == 1:
        return numeros[0] / 2

    # Caso 'Más de X', asumimos +10% del umbral
    if "mas de" in texto and len(numeros) == 1:
        return numeros[0] * 1.1

    return np.nan
```

```python
z_copy['E_VALORMATRICULAUNIVERSIDAD'] = z_copy['E_VALORMATRICULAUNIVERSIDAD'].apply(convertir_rango_a_numero)
```

```python
z_copy['E_VALORMATRICULAUNIVERSIDAD'].head(15)
```

Rellenar nulos

```python
moda = z_copy['E_VALORMATRICULAUNIVERSIDAD'].mode()[0]
z_copy['E_VALORMATRICULAUNIVERSIDAD'] = z_copy['E_VALORMATRICULAUNIVERSIDAD'].fillna(moda)
```

###Cleaning 'F_EDUCACIONMADRE' and 'F_EDUCACIONPADRE' column

```python
z_copy['F_EDUCACIONMADRE'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
import numpy as np

orden_educacion = [
    'sin info',
    'Primaria incompleta',
    'Primaria completa',
    'Secundaria (Bachillerato) incompleta',
    'Secundaria (Bachillerato) completa',
    'Técnica o tecnológica incompleta',
    'Técnica o tecnológica completa',
    'Educación profesional incompleta',
    'Educación profesional completa',
    'Postgrado'
]

# Reemplazar valores sin información and fill NaN
cols_to_clean = ['F_EDUCACIONMADRE', 'F_EDUCACIONPADRE']
for col in cols_to_clean:
    z_copy[col] = z_copy[col].replace(['Ninguno', 'No Aplica', 'No sabe'], 'sin info')
    z_copy[col] = z_copy[col].fillna('sin info')

# Creamos el codificador para ambas columnas, asegurando que 'sin info' sea manejado
# El parámetro 'categories' debe ser una lista de listas, una por cada columna
encoder = OrdinalEncoder(categories=[orden_educacion, orden_educacion])

# Aplicamos la codificación a ambas columnas al mismo tiempo

z_copy[cols_to_clean] = encoder.fit_transform(z_copy[cols_to_clean])
```

## Cleaning F_ESTRATOVIVIENDA column


```python
z_copy['F_ESTRATOVIVIENDA'] = (
    z_copy['F_ESTRATOVIVIENDA']
    .astype(str)
    .replace(r'Estrato\s*(\d+)', r'\1', regex=True)
    .replace('Sin Estrato', 0)
)

z_copy['F_ESTRATOVIVIENDA'] = pd.to_numeric(z_copy['F_ESTRATOVIVIENDA'], errors='coerce')

moda_estrato = z_copy['F_ESTRATOVIVIENDA'].mode()[0]
z_copy['F_ESTRATOVIVIENDA'] = z_copy['F_ESTRATOVIVIENDA'].fillna(moda_estrato)

z_copy['F_ESTRATOVIVIENDA'] = z_copy['F_ESTRATOVIVIENDA'].astype(int)
```

## Cleaning 'E_HORASSEMANATRABAJA' column
Mapeamos las 5 categorias existentes a valores numericos, posteriormente los valores nulos los reemplazamos

```python
z_copy['E_HORASSEMANATRABAJA'].unique()
```

```python
mapa_horas = {
    '0': 0,
    'Menos de 10 horas': 5,
    'Entre 11 y 20 horas': 15,
    'Entre 21 y 30 horas': 25,
    'Más de 30 horas': 35
}

z_copy['E_HORASSEMANATRABAJA'] = z_copy['E_HORASSEMANATRABAJA'].replace(mapa_horas)
z_copy['E_HORASSEMANATRABAJA'] = pd.to_numeric(z_copy['E_HORASSEMANATRABAJA'], errors='coerce')

mediana_horas = z_copy['E_HORASSEMANATRABAJA'].median()
z_copy['E_HORASSEMANATRABAJA'] = z_copy['E_HORASSEMANATRABAJA'].fillna(mediana_horas)
```
